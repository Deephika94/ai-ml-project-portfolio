### Multiclass Image classification using Convolutional neural network

### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image 
import os

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

import tensorflow as tf
import tensorflow.keras.layers as tfl
from tensorflow.keras.utils import to_categorical #for one hot encoding
```

### Load the dataset

```python
data = []
labels = []
classes = 43
print("current path:",os.getcwd())
```

#### Training data

```python
cur_path = "/Users/Deephika/Documents/2. Data science learning/5. AI-ML-project-portfolio/week13-CNN/data/"

for i in range(classes):
    path = os.path.join(cur_path, "train", str(i)) 
    images = os.listdir(path) # returns a list containing the names of the files in the path
    
    for image in images:
        try:
            im = Image.open(path + '/' + image)
            im = im.resize((30,30))
            im = np.array(im)
            data.append(im)
            labels.append(i)
        except:
            print("Error loading the image")

print(type(data))

# converting from List to Array
data = np.array(data)
labels = np.array(labels)
```

```python
print("Shape of data: ", data.shape)
print("Shape of labels: ",labels.shape)
```

#### Test data

```python
test = pd.read_csv(cur_path + "Test.csv")

labels = test["ClassId"].values
images = test["Path"].values

test_data = []

for image in images:
    try:
        im = Image.open(cur_path + '/' + image)
        im = im.resize((30,30))
        im = np.array(im)
        test_data.append(im)

    except:
        print("Error loading the image")

# converting from List to Array
test_data = np.array(test_data)

test_x = test_data/255
```

### Train and Validation split

```python
train_x, val_x, train_y, val_y = train_test_split(data,labels, test_size=0.2, random_state=42)

train_x = train_x/255
val_x = val_x/255

print("Train x shape: ", train_x.shape)
print("Train y shape: ", train_y.shape)
print("Val x shape: ", val_x.shape)
print("Val y shape: ", val_y.shape)
```

```python
# converting the labels into one hot encoding

train_y = to_categorical(train_y, classes)
val_y = to_categorical(val_y, classes)

print("Train y one hot encoded: ", train_y.shape)
print("Val y one hot encoded: ", val_y.shape)
```

```python
print(train_x[0].shape)
print(train_y[0])
```

### Convolutional neural network model

#### Model 1

```python
def cnn_model1(input_shape):
    
    model1 = tf.keras.Sequential([
     
        tfl.Input(shape = input_shape),
        
        tfl.Conv2D(filters=6, kernel_size=(5,5), strides=1),
        
        tfl.ReLU(),
        
        tfl.MaxPool2D(pool_size=(2,2), strides=2),
        
        tfl.Conv2D(filters=16, kernel_size=(5,5), strides=1),
        
        tfl.ReLU(),
        
        tfl.MaxPool2D(pool_size=(2,2), strides=2),
        
        tfl.Flatten(),
        
        tfl.Dense(units = 43, activation="softmax")  
        
    ]
    )
    
    return model1
```

```python
conv_model1 = cnn_model1((30, 30, 3))
conv_model1.compile(optimizer='adam',
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
conv_model1.summary()
```

```python
# Training the Model
with tf.device('/GPU:0'):
    epochs = 35
    history1 = conv_model1.fit(train_x, train_y, batch_size=128, epochs=epochs, validation_data=(val_x, val_y))
```

```python
print("Accuracy: ", history1.history["accuracy"][34])
print("Val accuracy: ", history1.history["val_accuracy"][34])
print("Loss: ", history1.history["loss"][34])
print("Val loss: ", history1.history["val_loss"][34])
```

```python
def plot_performance(history):
    df_loss_acc = pd.DataFrame(history.history)
    
    df_loss= df_loss_acc[['loss','val_loss']]
    df_loss.rename(columns={'loss':'train','val_loss':'validation'},inplace=True)
    
    df_acc= df_loss_acc[['accuracy','val_accuracy']]
    df_acc.rename(columns={'accuracy':'train','val_accuracy':'validation'},inplace=True)
    
    df_loss.plot(title='Model loss',figsize=(12,8)).set(xlabel='Epoch',ylabel='Loss')
    df_acc.plot(title='Model Accuracy',figsize=(12,8)).set(xlabel='Epoch',ylabel='Accuracy')
```

```python
plot_performance(history = history1)
```

#### Model 2

```python
def cnn_model2(input_shape):
    
    model2 = tf.keras.Sequential([
     
        tfl.Input(shape = input_shape),
        
        tfl.Conv2D(filters=32, kernel_size=(5,5), activation="relu"),
        
        tfl.Conv2D(filters=64, kernel_size=(5,5), activation="relu"),
        
        tfl.MaxPool2D(pool_size=(2,2)),
        
        tfl.Dropout(rate=0.15),
        
        tfl.Conv2D(filters=128, kernel_size=(3,3), activation="relu"),
        
        tfl.Conv2D(filters=256, kernel_size=(3,3), activation="relu"),
        
        tfl.MaxPool2D(pool_size=(2,2)),
        
        tfl.Dropout(rate=0.20),
        
        tfl.Flatten(),
        
        tfl.Dense(units = 512, activation="relu"),
        
        tfl.Dropout(rate=0.25),
        
        tfl.Dense(units = 43, activation="softmax")  
        
    ]
    )
    
    return model2
```

```python
conv_model2 = cnn_model2((30,30,3))
conv_model2.compile(loss = "categorical_crossentropy",
                           optimizer = "adam",
                           metrics=['accuracy'])
conv_model2.summary()
```

```python
with tf.device('/GPU:0'):
    history2 = conv_model2.fit(train_x, train_y, batch_size=128, epochs=35, validation_data=(val_x,val_y))
```

```python
plot_performance(history = history2)
```

```python
print("Accuracy: ", history2.history["accuracy"][34])
print("Val accuracy: ", history2.history["val_accuracy"][34])
print("Loss: ", history2.history["loss"][34])
print("Val loss: ", history2.history["val_loss"][34])
```

### Predicting the test dataset

```python
y_pred = np.argmax(conv_model2.predict(test_x), axis=-1)

print("Test data accuracy:", accuracy_score(labels, y_pred)*100)
```
